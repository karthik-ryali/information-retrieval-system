# Text Compression — Run-Length Encoding (RLE)

**Goal:** Compress a string by collapsing consecutive repeated characters into `character + count`.

---

## Theory

RLE replaces runs of the same character with a compact representation:

```
Original  : A A G G T T T 6 6
Compressed: A2 G2 T3 62
```

It is **lossless** — the original can be perfectly recovered.

| Scenario | Result |
|----------|--------|
| Long repeated runs (`AAAAAAA`) | Significant compression |
| Random text (`ABCDEFG`) | May actually expand (each char gets a count digit) |

**Real-world uses:** DNA sequences, pixel runs in images (BMP), fax transmission

````python
def rle_compress(text: str) -> str:
    """
    Compress text using Run-Length Encoding.
    Consecutive repeated chars → char + count.

    'AAGGTTT66' → 'A2G2T362'
    """
    if not text:
        return ""

    compressed = []
    count = 1

    for i in range(1, len(text)):
        if text[i] == text[i - 1]:
            count += 1
        else:
            compressed.append(text[i - 1] + str(count))
            count = 1

    compressed.append(text[-1] + str(count))   # flush the last run
    return "".join(compressed)


def rle_decompress(compressed: str) -> str:
    """
    Reverse RLE back to the original string.
    Handles multi-digit counts (e.g., A12 → 'AAAAAAAAAAAA').
    """
    result = []
    i = 0

    while i < len(compressed):
        char = compressed[i]
        i += 1

        num_str = ""
        while i < len(compressed) and compressed[i].isdigit():
            num_str += compressed[i]
            i += 1

        result.append(char * int(num_str))

    return "".join(result)
````

````python
# ── Demo: compress, decompress, check ratio ────────────────
samples = [
    "AAGGTTT66",          # original test case
    "AAABBBCCCC",         # clear runs
    "ABCDE",              # no repeats — watch the ratio
    "AAAAAAAAAA",         # all same — best case for RLE
]

print(f"{'Original':<18} {'Compressed':<15} {'Recovered':<18} {'Ratio':>7}")
print("-" * 60)

for text in samples:
    comp   = rle_compress(text)
    decomp = rle_decompress(comp)
    ratio  = len(comp) / len(text)
    check  = "✔" if decomp == text else "✘ MISMATCH"

    print(f"{text:<18} {comp:<15} {decomp:<18} {ratio:>6.2f}x  {check}")
````
