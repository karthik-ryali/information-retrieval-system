# Text Pattern Matching — Brute Force Search

**Goal:** Locate all positions where a short string (pattern) appears inside a longer string — implemented from scratch.

---

## How It Works

This is a **sliding window** approach:
1. Start at index `0` of the main string
2. Compare `pattern` character-by-character starting from that index
3. If all characters match → record the position
4. Slide one step to the right and repeat

```
main:    A A B A A C A A D A A B A A B A
pattern: A A B A
         ↑ check here → match at index 0

         shift →
           A A B A → no match at index 1
         ...
                             A A B A → match at index 9
```

**Time Complexity:** O(n × m) — n = len(main), m = len(pattern)  
For large inputs, use **KMP** or **Boyer-Moore** which run in O(n + m).

````python
def locate(pattern: str, main: str) -> None:
    """
    Print all positions where `pattern` appears in `main`.
    Uses brute-force sliding window comparison.
    """
    n = len(main)
    m = len(pattern)
    found = False

    print(f"Main String  : {main}")
    print(f"Pattern      : {pattern}")
    print("-" * 45)

    for i in range(n - m + 1):           # slide across main
        for k in range(m):               # compare char by char
            if main[i + k] != pattern[k]:
                break
        else:
            # Python for-else: 'else' runs only if loop was NOT broken
            print(f"  ✔ Match at index: {i}")
            found = True

    if not found:
        print("  Pattern NOT found.")


# ── Tests ──────────────────────────────────────────────────
print("Test 1:")
locate("AABA", "AABAACAADAABAABA")

print()
print("Test 2 — overlapping pattern:")
locate("AA", "AAAA")

print()
print("Test 3 — not found:")
locate("ZZZ", "AABAACAAD")
````

## Key Concepts

| Term | Meaning |
|------|---------|
| Sliding window | Moving a fixed-size view across the string |
| for-else | Python's `else` on a loop runs if the loop completed without `break` |
| O(n × m) | Each of n positions checks up to m characters |
